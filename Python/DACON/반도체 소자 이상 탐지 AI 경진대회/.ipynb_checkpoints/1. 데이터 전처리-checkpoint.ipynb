## 1. 데이터 전처리

```python
from torchvision import transforms

# 이미지 전처리 및 변환
preprocess = transforms.Compose([
    transforms.Resize((224, 224)),  # 이미지 크기를 조정합니다. (224x224 크기로 조정됩니다)
    transforms.ToTensor(),          # 이미지를 PyTorch 텐서로 변환합니다.
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # 이미지를 정규화합니다.(평균 0.5, 표준 편차 0.5로 정규화합니다)
])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_moons

# 데이터 생성 (무작위로 생성된 2차원 데이터셋)
X, _ = make_moons(n_samples=200, noise=0.1, random_state=42)

# One-Class SVM 모델 초기화
svm_model = svm.OneClassSVM(nu=0.05, kernel='rbf', gamma=0.1)

# 모델 학습 (정상 데이터만 사용)
svm_model.fit(X)

# 예측 (이상치 탐지)
y_pred = svm_model.predict(X)

# 이상치와 정상 데이터를 구분하여 시각화
plt.figure(figsize=(8, 6))

# 이상치 표시 (파란색)
plt.scatter(X[y_pred == -1][:, 0], X[y_pred == -1][:, 1], color='blue', label='Outliers')

# 정상 데이터 표시 (노란색)
plt.scatter(X[y_pred == 1][:, 0], X[y_pred == 1][:, 1], color='yellow', label='Inliers')

plt.title('One-Class SVM 이상치 탐지 결과')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
